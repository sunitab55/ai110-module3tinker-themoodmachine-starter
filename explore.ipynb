```python
from mood_analyzer import MoodAnalyzer
from dataset import SAMPLE_POSTS, TRUE_LABELS

analyzer = MoodAnalyzer()

for text, true_label in zip(SAMPLE_POSTS, TRUE_LABELS):
    predicted = analyzer.predict_label(text)
    score = analyzer.score_text(text)
    tokens = analyzer.preprocess(text)
    print(f"Post:      {text}")
    print(f"Tokens:    {tokens}")
    print(f"Score:     {score}")
    print(f"Predicted: {predicted}  |  True: {true_label}  |  {'✓' if predicted == true_label else '✗'}")
    print()
```
